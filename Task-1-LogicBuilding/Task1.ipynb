```python
#1. User Login Check


# Given credentials
username = "admin"
password = "1234"

# Input credentials (example)
input_username = "admin"
input_password = "1234"

# Authentication check
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer

# Given marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final result
print("Student Result Analysis")
print("-----------------------")
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#3. Simple Data Cleaner

# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Clean and standardize names
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()  # remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer

# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze message sizes
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters")
```

```python
#5. Error Message Detector

# Given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

# Print result
print("Total ERROR entries:", error_count)
```
