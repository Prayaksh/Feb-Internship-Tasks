```python
# 1. 
def smart_parking_system(capacity, vehicle_logs):
    
    current_parked = 0
    peak_usage = 0
    
    # Loop through logs
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1
        
        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked
    
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    
    # Capacity alert
    if current_parked > capacity:
        print("Parking Status: Capacity Exceeded - Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
# 2. 
def estimate_delivery_time(distance, traffic_level, weather):
    
    # Base time calculation
    base_time = distance * 5
    
    # Traffic delay
    if traffic_level == "Low":
        traffic_delay = 5
    elif traffic_level == "Medium":
        traffic_delay = 10
    else:  # High
        traffic_delay = 15
    
    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    else:  # Rainy
        weather_delay = 10
    
    # Final ETA
    final_eta = base_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", final_eta, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# 3. 
def analyze_seat_occupancy(total_seats, booked_seats):
    
    booked_count = 0
    
    # Loop to count booked seats
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    
    # Suggest additional show
    if occupancy >= 90:
        print("Suggestion: Consider Opening Additional Show")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

analyze_seat_occupancy(total_seats, booked_seats)
```

```python
# 4.
def classify_server_load(cpu_readings):
    
    total = 0
    
    # Loop to calculate total CPU usage
    for reading in cpu_readings:
        total += reading
    
    average_cpu = total / len(cpu_readings)
    
    print("Average CPU Load:", int(average_cpu), "%")
    
    # Classification
    if average_cpu < 50:
        print("Server Status: Normal")
    elif average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

```python
# 5. 
def monitor_classroom_resources(resource_usage):
    
    overused_resources = []
    
    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)
    
    # Output results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_classroom_resources(usage_data)
```

```python
# 6.
def manage_event_registrations(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    
    # Loop to simulate registrations
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
manage_event_registrations(100, 105)
```
