```python
#1. Unique Words in a Sentence

# Take input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#2. Highest Salary from Employee Data

# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Print result
print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")
```

```python
#3. Find Maximum and Minimum Values

# Input list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_val = numbers[0]
min_val = numbers[0]

# Loop through list
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

# Print results
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
#4. Count Products Above a Price Threshold

# Input list
prices = [450, 1200, 899, 1500, 300]

# Counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
#5. Calculate Attendance Percentage

# Input list
attendance = ["P", "P", "A", "P", "P"]

# Count number of presents
present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

# Print result
print("Attendance Percentage:", percentage)
```

```python
#6. Remove Duplicate Phone Numbers

# Input list
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

```python
#7. Count Character Frequency

# Input string
text = "pythonp"

# Dictionary to store frequency
char_count = {}

# Loop through characters
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Print result
print(char_count)
```

```python
#8. Convert List to Tuple

# Input list
numbers = [10, 20, 30]

# Convert list to tuple
numbers_tuple = tuple(numbers)

# Print result
print("Tuple:", numbers_tuple)
```

```python
#9. Check if a key exists in a Dictionary

#Example dictionary
employees = {
    "Employee": "Ravi",
    "Salary": 75000
}

# Key to check
key_to_check = "Employee"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#10. Calculate Average Marks

# Input list
marks = [65, 70, 80, 75]

# Calculate average
average = sum(marks) / len(marks)

# Print result
print("Average Marks:", average)
```
