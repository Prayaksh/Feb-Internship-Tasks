```python
# 1.
# Employee performance data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
# 2.
import string

# Input query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
# 3.
# Sensor readings input
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Find even readings and store as (hour_index, reading_value)
valid_readings = [(index, value) 
                  for index, value in enumerate(sensor_readings) 
                  if value % 2 == 0]

# Step 2: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# 4. 
# Email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domain usage
domain_count = {}

for email in emails:
    domain = email.split('@')[1]   # extract domain
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# 5.
# Sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect spikes (more than 30% above average)
threshold = average_sales * 1.30

# Step 3: Display day number and sale value
for index, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {index}: {value}")
```

```python
# 6.
# User IDs input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
